```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
users = pd.read_csv('users.csv')
photos = pd.read_csv('photos.csv')
likes = pd.read_csv('likes.csv')
comments = pd.read_csv('comments.csv')
follows = pd.read_csv('follows.csv')
tags = pd.read_csv('tags.csv')
photo_tags = pd.read_csv('photo_tags.csv')
```

```python
print(users.head())
print(users.shape)
print(photos.shape)
print(likes.shape)
print(comments.shape)
```

```python
print("Users")
print(users.isnull().sum())

print("Photos")
print(photos.isnull().sum())

print("Likes")
print(likes.isnull().sum())

print("Comments")
print(comments.isnull().sum())
```

```python
print(users.dtypes)
print(photos.dtypes)
print(likes.dtypes)
```

```python
import matplotlib.pyplot as plt

top_tags = photo_tags.merge(
    tags,
    left_on='tag ID',
    right_on='id'
)

top_tags['tag text'].value_counts().head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 Most Used Hashtags')
plt.xlabel('Hashtags')
plt.ylabel('Frequency')
plt.show()
```

```python
photos['user ID'].value_counts().head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top Active Users')
plt.xlabel('User ID')
plt.ylabel('Number of Posts')
plt.show()
```

```python
top_likes = likes.groupby('photo').size()

top_likes.sort_values(
    ascending=False
).head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 Most Liked Posts')
plt.xlabel('Photo ID')
plt.ylabel('Likes')
plt.show()
```

```python
top_followed = follows['followee '].value_counts()

top_followed.head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Most Followed Users')
plt.xlabel('User ID')
plt.ylabel('Followers')
plt.show()
```

```python
users['Verified status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title('Verified vs Non-Verified Users')
plt.show()
```

```python
users['private/public'].value_counts().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Public vs Private Accounts')
plt.xlabel('Account Type')
plt.ylabel('Number of Users')
plt.show()
```

```python
photos['Insta filter used'].value_counts().head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top Instagram Filters')
plt.xlabel('Filter')
plt.ylabel('Usage Count')
plt.show()
```
